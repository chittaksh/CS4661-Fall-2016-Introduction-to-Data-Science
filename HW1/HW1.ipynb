### a) We have to read the iris dataset from a url and save it into Pandas DataFrame, Inorder to proceed we have to folow the steps:

```python
# Since we have to read the iris dataset from  following URL:'https://raw.githubusercontent.com/mpourhoma/CS4661/master/iris.csv'
# and assign it to a Pandas DataFrame 

# we first import Panda library 
import pandas as pd

# read the iris dataset and store it in panda dataframe
iris_dataf = pd.read_csv('https://raw.githubusercontent.com/mpourhoma/CS4661/master/iris.csv')

# Display the Data Frame 
iris_dataf
```

### b) Split the	 dataset	 into	 testing	 and	 training sets	 with	 the	 following	 parameters:
### test_size=0.4,	random_state=1.
#### 1. Import KNeighborsClassifier "Class".
#### 2. Running the sklearn funcition load_iris() to instantiate an "object" containing iris datset.
#### 3. Store the values in X and y .
#### 4. Split the dataset in training and testing dataset.

```python
#Split the dataset into testing and training sets with the following parameters:
#test_size=0.4, random_state=1

# import KNeighborsClassifier "Class"
from sklearn.neighbors import KNeighborsClassifier

#load the data from sklearn .datasets library , as it has inbuilt iris data set
from sklearn.datasets import load_iris

iris = load_iris()
#print(iris)

#"data" attribute will return the iris dataset features:
X= iris.data

# "target" attribute will return the iris dataset labels 
y=iris.target

#Split the dataset in training and testing dataset.
from sklearn.cross_validation import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X, y, test_size=0.4, random_state=1)
```

#### Printing the size of X and y.

```python
#print the size of the training set
print(X_train.shape)
print(y_train.shape)
```

```python
#print the size of the testing set
print(X_test.shape)
print(y_test.shape)
print(y_test)
```

### c) Instantiate	a	KNN	object	with	K=3,	train	it	on	the	training	set	and	test	it	on	the	testing	set.	
### Then,	calculate	the	accuracy the prediction.
#### 1. Instantiating an "object" of KNeighborsClassifier "class" with k=3
#### 2. We use the method "fit" of the object along with training dataset and labels to train the model.
#### 3. Calculate the y_predict so as to get the accuracy between our predicted and testing data.
#### 4. Import accuracy_score from sklearn .metrics library, so as to calculate the accuracy of our prediction


```python
#Instantiate a KNN object with K=3, train it on the training set and test it on the testing set.
#Then, calculate the accuracy of your prediction

# Instantiating an "object" of KNeighborsClassifier "class" with k=3:

k=3

knn_for_iris = KNeighborsClassifier(n_neighbors=k)
```

```python
#Traning can only be done on training data set

knn_for_iris.fit(X_train, y_train)
```

```python
#Testing can be done on testing data set

y_predict = knn_for_iris.predict(X_test)
print(y_predict)
```

```python
#Accuracy Evaluation
from sklearn.metrics import accuracy_score 

score = accuracy_score(y_test, y_predict)
print(score)
```

### d) Repeat	part	(c)		for	K=1,	K=5,	K=7,	K=15,	K=28,	K=60.
#### 1. We creat a list that consists of all the values of K's as list1 .
#### 2. Then I run a for loop on the list1 and calculate the accuracy for every value of k.
#### 3. Then I append all the accuracy score in the list named as list2.

```python
#Repeat part (c) for K=1, K=5, K=7, K=15, K=28, K=60 
list1 = [1, 5, 7, 15, 28, 60]
list2=[]
for ks in list1:
    
    
    # Instantiating an "object" of KNeighborsClassifier "class" with k=ks:
    knn_for_iris_ks = KNeighborsClassifier(n_neighbors=ks)
    
    #Split the dataset in training and testing dataset.
    X_train, X_test,y_train,y_test = train_test_split(X, y, test_size=0.4, random_state=1)
    
    #Traning can only be done on training data set
    knn_for_iris_ks.fit(X_train, y_train)
    
    #Testing can be done on testing data set
    y_predict_ks = knn_for_iris_ks.predict(X_test)
    
    #Accuracy Evaluation
    score = accuracy_score(y_test, y_predict_ks)
    list2.append(score)
    
print(list2)
```

#### First the accuracy_score increases and then  becomes constant  ove the values of k and then starts decreacing.

### e) Make	prediction	based on only ONE single feature
#### 1. Stored all the features in a list called feature_cols.
#### 2. Instantiating an "object" of KNeighborsClassifier "class" with k=3.
#### 3. Run a for loop so as to calculate the accuracy of every feature seperately (repeating all the steps as we did in questin c)
#### 4. Append the accuracy_scores  of every feature in a list called list 2.


```python
# make prediction based on only ONE single feature
def categorical_to_numeric(x):
    if x == 'setosa':
        return 0
    elif x == 'versicolor':
        return 1
    elif x == 'virginica':
        return 2
    
iris_dataf['label'] = iris_dataf['species'].apply(categorical_to_numeric)

feature_cols = ['sepal_length','sepal_width','petal_length','petal_width']
list2=[]


k=3
knn_for_iris = KNeighborsClassifier(n_neighbors=k)
#print(A_train.shape)
#print(b_train.shape)

for f in feature_cols:
    A= iris_dataf[[f]]
    b= iris_dataf['label']
    A_train, A_test,b_train,b_test = train_test_split(A, b, test_size=0.4, random_state=1)
    #print(A,b)
    knn_for_iris.fit(A_train, b_train)
    b_predict = knn_for_iris.predict(A_test)
    score = accuracy_score(b_test, b_predict)
    list2.append(score)
    
print(list2)
```

#### The most accurate score is of petal_length feature and the second best is petal width .
