## LOGISTIC REGRESSION and DECISION TREE CLASSIFIERS:

### Que 1 - Read the dataset file “Hearts_s.csv” (you should download it from CSNS), and assign it to a Pandas DataFrame.

```python
# Importing the required packages and libraries
import pandas as pd

# The following line will import LogisticRegression and DecisionTreeClassifier Classes
# "LogisticRegression" and "DecisionTreeClassifier" are the names of a "sklearn classes" to perform Logistic Regression and Decision Tree based Classification 
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.neighbors import KNeighborsClassifier
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score 

#Importing the data from the source file.
heart_df = pd.read_csv('http://www.immensity.in/Heart_s.csv')

#Testing if read the file.
heart_df[0::10]
```

### Que 2 - Check out the dataset. As you see, there are a number of features, which include both contextual and biological factors (e.g. age, gender, vital signs, …). The last column “AHD” is the label with “Yes” meaning that a human subject has Heart Disease, and “No” meaning that the subject does not have Heart Disease.

```python
## Code Que 2 here
def categorical_to_numeric(x):
    if x == 'Yes':
        return 1
    elif x == 'No':
        return 0
   
    
heart_df['label'] = heart_df['AHD'].apply(categorical_to_numeric)
y= heart_df['label']
y
```

### Que 3 - As you see, there are at least 3 categorical features in the dataset (Gender, ChestPain, Thal). Let’s ignore these categorical features for now, only keep the numerical features and build your feature matrix and label vector.

```python
## code que 3 here
feature_cols = ['Age','RestBP','Chol','RestECG','MaxHR','Oldpeak']
#heart_dataf[feature_cols]

#Normalization of the data.
for f  in feature_cols:
    max_feature = heart_df[f].max()
    heart_df[f]= heart_df[f]/max_feature
    #print(max_feature)
#heart_dataf

X= heart_df[feature_cols]
X
```

### Que 4 - Split the dataset into testing and training sets with the following parameters: test_size=0.3, random_state=3.

```python
## Code que 4 here
X_train, X_test,y_train,y_test = train_test_split(X, y, test_size=0.3, random_state=3)
```

### Que 5 - Use KNN (with k=5), Decision Tree, and Logistic Regression Classifiers to predict Heart Disease based on the training/testing datasets that you built in part (d). Then check, compare, and report the accuracy of these 3 classifiers. Which one is the best? Which one is the worst?

```python
## Code Que 5 here.
#KNN
k=5
knn_for_heart = KNeighborsClassifier(n_neighbors=k)

#knn_for_heart.fit(X,y)

knn_for_heart.fit(X_train, y_train)
y_predict = knn_for_heart.predict(X_test)
score = accuracy_score(y_test, y_predict)
print(score)
```

```python
#Logistic Regression
my_logreg = LogisticRegression()

#my_logreg.fit(X,y)

my_logreg.fit(X_train, y_train)
y_predict_lr = my_logreg.predict(X_test)
score_lr = accuracy_score(y_test, y_predict_lr)

print(score_lr)
```

```python
#Decision Tree 
my_decisiontree = DecisionTreeClassifier()

#my_decisiontree.fit(X, y)

my_decisiontree.fit(X_train, y_train)
y_predict_dt = my_decisiontree.predict(X_test)
score_dt = accuracy_score(y_test, y_predict_dt)

print(score_dt)
```

### Que 6 - Now, we want to use the categorical features as well! Thus, we have to perform OneHotEncoding for the categorical features. To do this, each categorical feature should be replaced with dummy columns in the feature table (one column for each possible value of a categorical feature), and then encode it in a binary manner such that at most only one of the dummy columns can take “1” at a time (and zero for the rest). For example, “Gender” can take two values “m” and “f”. Thus, we need to replace this feature (in the feature table) by two columns titled “m” and “f”. Wherever, we have a male subject, we can put “1” and ”0” in the columns “m” and “f”. Wherever, we have a female subject, we can put “0” and ”1” in the columns “m” and “f”. (Hint: you will need 4 columns to encode “ChestPain” and 3 columns to encode “Thal”).

```python
# Genereating two new columns for Gender and deleteing the original.
heart_df['male'] = heart_df['Gender'].map(lambda x: 1 if x=='m' else 0)
heart_df['female'] = heart_df['Gender'].map(lambda x: 1 if x=='f' else 0)
#heart_df = heart_df.drop('Gender',1)

# Genereating four new columns for ChestPain and deleteing the original.
heart_df['typical'] = heart_df['ChestPain'].map(lambda x: 1 if x=='typical' else 0)
heart_df['asymptomatic'] = heart_df['ChestPain'].map(lambda x: 1 if x=='asymptomatic' else 0)
heart_df['nontypical'] = heart_df['ChestPain'].map(lambda x: 1 if x=='nontypical' else 0)
heart_df['nonanginal'] = heart_df['ChestPain'].map(lambda x: 1 if x=='nonanginal' else 0)
#heart_df = heart_df.drop('ChestPain',1)

# Genereating three new columns for Thal and deleteing the original.
heart_df['fixed'] = heart_df['Thal'].map(lambda x: 1 if x=='fixed' else 0)
heart_df['normal'] = heart_df['Thal'].map(lambda x: 1 if x=='normal' else 0)
heart_df['reversable'] = heart_df['Thal'].map(lambda x: 1 if x=='reversable' else 0)
#heart_df = heart_df.drop('Thal',1)

heart_df[0::10]
```
